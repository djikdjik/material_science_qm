\*\*\*Exercise 2.4. Using the same approximations introduced in Exercise 2.2, show that the ground\-state electron density, n\(r\), of the He atom is given by:

$$
n(r) =\frac{16}{\pi}\exp(−4r)\; \cdots\;(2.45)
$$

We know that for each electron,

$$
\phi(r)_{1s}=\frac{2^{3/2}}{\sqrt{\pi}} \exp(-2r)
$$

One can proceed, using Eq$2.44$ to get;

$$
n(r)=|\phi_1|^2+|\phi_2|^2=\frac{2^4}{\pi}\exp(-4r)
$$



\*\*\*Check that the integral of the density correctly yields two electrons. One needs to integrate over the whole space, taking the spherical approach which seems easier.


```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import sympy as smp
```

```python
r, phi, theta = smp.symbols('r \phi \theta')
n = smp.symbols('n' , cls=smp.Function)
n = 16/smp.pi * smp.exp(-4*r)
```

```python
smp.integrate(n*r**2*smp.sin(theta),(r, 0, smp.oo),(theta, 0, smp.pi), (phi, 0, 2*smp.pi))
```

This is as expected since the density integratl should yield total number of electrons, which in this case is 2.



\*\*\*Plot the radial probability, 4πr 2 n\(r\), of finding any electron at a distance $r$ and determine the distance at which this probability reaches its maximum.



```python
r = np.arange(0, 3.5, 0.01)
plt.plot(r, 4*np.pi*r**2* 16/np.pi*np.exp(-4*r), color = 'b') 

plt.xlabel('r', fontsize = 16) 
plt.ylabel('probability', fontsize = 16) 
plt.grid()
plt.title('radial probabilities', fontsize = 16)
plt.show()
```

Though the peak ocurs at 0.5, one can still confirm as follows.



```python
from sympy import Eq, solve, simplify

dnpdr = smp.diff(4*np.pi*r**2*n, r)

eq = Eq(0, dnpdr)
print(solve(eq, r))
print(solve(simplify(eq), r))
```
